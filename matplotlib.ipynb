```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('시군구별_이동자수_20220216221103.xlsx',header = 0)
```

```python
df.isna().sum()
```

```python
df
```

```python
df_year = df.columns[1:]
```

```python
df_seoul = df.values[2][1:]
df_busan = df.values[3][1:]
df_degu = df.values[4][1:]
df_incheon = df.values[5][1:]
```

```python
df_seoul
```

```python
df_year 
```

```python
plt.plot(df_year,df_seoul)
```

```python
## 차트 이름 제목 추가
## 한글 폰트
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'

plt.plot(df_year,df_seoul)
plt.title('서울시 전입 인구')
plt.xlabel('기간')
```

```python
## 글자 겹치지 않게 하기
plt.figure(figsize=(14,5))
plt.plot(df_year,df_seoul)

## 글자 회전
plt.xticks(rotation='vertical')
plt.plot(df_year,df_seoul)
plt.title('서울시 전입 인구')
plt.xlabel('기간')
```

```python
## 스타일 서식
plt.style.use('ggplot')

## 마커 표시
plt.figure(figsize=(14,5))
plt.plot(df_year,df_seoul,marker='o', markersize = 5)
plt.xticks(rotation='vertical')
plt.title('서울시 전입 인구')
plt.xlabel('기간')
## label
plt.legend(labels = ['서울 전입 인구'],loc='best')
```

```python
### 여러항목 표시
## 스타일 서식
plt.style.use('ggplot')

## 마커 표시
plt.figure(figsize=(14,5))
plt.plot(df_year,df_seoul,marker='o', markersize = 5,label = '서울')
plt.plot(df_year,df_busan,marker='o', markersize = 5,label = '부산')
plt.plot(df_year,df_degu,marker='o', markersize = 5,label = '대구')
plt.plot(df_year,df_incheon,marker='o', markersize = 5,label = '인천')
plt.xticks(rotation='vertical')
plt.title('광역시 전입 인구')
plt.xlabel('기간')
## label
plt.legend(loc="best")
```
