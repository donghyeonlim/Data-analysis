```python
## 차트 이름 제목 추가
## 한글 폰트
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'

plt.plot(df_year,df_seoul)
plt.title('서울시 전입 인구')
plt.xlabel('기간')
```

```python
## 글자 겹치지 않게 하기
plt.figure(figsize=(14,5))
plt.plot(df_year,df_seoul)

## 글자 회전
plt.xticks(rotation='vertical')
plt.plot(df_year,df_seoul)
plt.title('서울시 전입 인구')
plt.xlabel('기간')
```

```python
## 스타일 서식
plt.style.use('ggplot')

## 마커 표시
plt.figure(figsize=(14,5))
plt.plot(df_year,df_seoul,marker='o', markersize = 5)
plt.xticks(rotation='vertical')
plt.title('서울시 전입 인구')
plt.xlabel('기간')
## label
plt.legend(labels = ['서울 전입 인구'],loc='best')
```
