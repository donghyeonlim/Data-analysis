```python
## 차트 이름 제목 추가
## 한글 폰트
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'

plt.plot(df_year,df_seoul)
plt.title('서울시 전입 인구')
plt.xlabel('기간')
```

```python
## 글자 겹치지 않게 하기
plt.figure(figsize=(14,5))
plt.plot(df_year,df_seoul)

## 글자 회전
plt.xticks(rotation='vertical')
plt.plot(df_year,df_seoul)
plt.title('서울시 전입 인구')
plt.xlabel('기간')
```

```python
## 스타일 서식
plt.style.use('ggplot')

## 마커 표시
plt.figure(figsize=(14,5))
plt.plot(df_year,df_seoul,marker='o', markersize = 5)
plt.xticks(rotation='vertical')
plt.title('서울시 전입 인구')
plt.xlabel('기간')
## label
plt.legend(labels = ['서울 전입 인구'],loc='best')
```

```python
### 여러항목 표시
## 스타일 서식
plt.style.use('ggplot')

## 마커 표시
plt.figure(figsize=(14,5))
plt.plot(df_year,df_seoul,marker='o', markersize = 5,label = '서울')
plt.plot(df_year,df_busan,marker='o', markersize = 5,label = '부산')
plt.plot(df_year,df_degu,marker='o', markersize = 5,label = '대구')
plt.plot(df_year,df_incheon,marker='o', markersize = 5,label = '인천')
plt.xticks(rotation='vertical')
plt.title('광역시 전입 인구')
plt.xlabel('기간')
## label
plt.legend(loc="best")
```

```python
from tensorflow import keras
dataset_path = keras.utils.get_file("auto-mpg.data", "http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data")
```

```python
column_names = ['MPG','Cylinders','Displacement','Horsepower','Weight',
                'Acceleration', 'Model Year', 'Origin']
raw_dataset = pd.read_csv(dataset_path, names=column_names,
                      na_values = "?", comment='\t',
                      sep=" ", skipinitialspace=True)

dataset = raw_dataset.copy()
```

```python
dataset
```

히스토그램

```python
plt.rcParams['font.family'] = 'gulim.ttc'
plt.rcParams['axes.unicode_minus'] = False
plt.style.use('classic')
dataset['MPG'].plot(kind='hist',bins=10,color='orange',figsize=(10,5))
plt.title('hist')
plt.xlabel('MPG')
```
